# Завдання 2. Робота з netflix_titles.csv

Робота з датасетом Netflix titles для фільтрації та обробки даних.


```python
import pandas as pd

netflix = pd.read_csv('netflix.csv', parse_dates=['date_added'])

netflix.info()
```

```python
## 2) Знайти усі рядки з назвою Limitless

limitless_rows = netflix[netflix['title'] == 'Limitless']
limitless_rows
```

```python

## 3) Знайти всі рядки з режисером Robert Rodriguez та типом Movie
robert_rodriguez_movies = netflix[(netflix['director'] == 'Robert Rodriguez') & (netflix['type'] == 'Movie')]
robert_rodriguez_movies
```

```python
## 4) Знайти усі рядки з датою початку трансляції на Netflix 2019-07-31 або режисером Robert Altman
target_date = pd.to_datetime('2019-07-31')
robert_altman_or_date = netflix[(netflix['date_added'] == target_date) | (netflix['director'] == 'Robert Altman')]
robert_altman_or_date
```

```python

## 5) Знайти усі рядки з режисером Orson Welles, Aditya Kripalani та Sam Raimi (без використання оператора OR)
directors_list = ['Orson Welles', 'Aditya Kripalani', 'Sam Raimi']
directors_filter = netflix[netflix['director'].isin(directors_list)]
directors_filter
```

```python

## 6) Знайти усі рядки зі значеннями стовпця date_added між 1 травня та 1 липня 2019 (без використання оператора AND)
start_date = pd.to_datetime('2019-05-01')
end_date = pd.to_datetime('2019-07-01')
date_range_filter = netflix[netflix['date_added'].between(start_date, end_date, inclusive='left')]
date_range_filter
```

```python

## 7) Видалити всі рядки з NaN в стовпці director
netflix_clean = netflix.dropna(subset=['director'])

netflix_clean.head()
```

```python
date_counts = netflix['date_added'].value_counts()

single_movie_dates = date_counts[date_counts == 1].index

single_movie_days = netflix[netflix['date_added'].isin(single_movie_dates)]
single_movie_days
```
