# Завдання 1. Робота з об'єктами DataFrame

Файл nfl.csv містить список гравців Національної ліги з наступними стовпцями: Name, Team, Position, Birthday та Salary.


```python
import pandas as pd

nfl = pd.read_csv('nfl.csv', parse_dates=['Birthday'], index_col='Name')

print(nfl)
```

```python
## 2) Маніпуляції з даними

team_capacity = nfl['Team'].value_counts()

print("Перші три елементи:")
print(team_capacity.head(3))

print("\nОстанні три елементи:")
print(team_capacity.tail(3))
```

```python
team_with_most_players = team_capacity.index[0]

players_from_largest_team = nfl[nfl['Team'] == team_with_most_players]

oldest_highest_paid = players_from_largest_team.sort_values(['Birthday', 'Salary'], ascending=[True, False]).iloc[0]

print("Найбільшоплачуваний старший гравець команди з найбільшою чисельністю гравців:")
print(oldest_highest_paid[['Team', 'Position', 'Birthday', 'Salary']])
```

```python
team_with_least_players = team_capacity.index[-1]

players_from_smallest_team = nfl[nfl['Team'] == team_with_least_players]

youngest_highest_paid = players_from_smallest_team.sort_values(['Birthday', 'Salary'], ascending=[False, False]).iloc[0]

print("Найбільшоплачуваний молодший гравець команди з найменшою чисельністю гравців:")
print(youngest_highest_paid[['Team', 'Position', 'Birthday', 'Salary']])
```

```python
## 3) Додаткові маніпуляції з даними

nfl = nfl.reset_index()
nfl = nfl.set_index('Team')

nfl.head()
```

```python
nfl = nfl.rename(columns={'Birthday': 'Date of Birth'})

nfl.head()
```
